```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
losses = pd.read_csv('results/losses_11_01_2023__21_16_44.csv')
test_metrics = pd.read_csv('results/test_metrics_11_01_2023__21_16_44.csv')
```

```python
test_metrics
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

for l in ['BinaryAccuracy', 'BinaryPrecision', 'BinaryRecall', 'BinaryF1Score', 'BinaryMatthewsCorrCoef']:
    ax.plot(test_metrics['epoch'], test_metrics[l], label=l)
    ax.legend()
    ax.grid(True)
    ax.set(xlabel='Epoch', ylabel="Score", title='Test metrics')
fig.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(range(len(losses)), losses['BCE loss'], label='BCE Loss')
ax.legend()
ax.grid(True)
ax.set(xlabel='Epoch', ylabel="Loss", title='BCE Loss')
fig.tight_layout()
```
